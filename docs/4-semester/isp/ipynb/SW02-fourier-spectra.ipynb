## **&#9734;  Exercise 1 – Spectrum**

The **amplitude spectrum** of a signal describes how strongly the individual harmonics (integer multiples of the fundamental frequency 
$f=\tfrac{1}{T}$) contribute to the signal. It reveals *how the signal's energy (or amplitude) is distributed across frequencies* in the frequency domain.

The **phase spectrum**, on the other hand, specifies the relative phase shifts of these harmonics. These phase shifts determine how the sinusoidal components align in time and, consequently, *how the waveform is shaped in the time domain*.

In the previous notebook, we computed the Fourier coefficients of periodic signals (rectangular, sawtooth, and triangular waves) using the sine–cosine form. To obtain the magnitude (amplitude) and phase representation of the spectrum, we use the following conversion formulas:

$$
\begin{align*}
A_k &= \sqrt{a_k^2 + b_k^2} \\[0.5em]
\varphi_k&=\arctan{\left(\frac{b_k}{a_k}\right)}
\end{align*}
$$

As the number of terms $N$ increases, the approximation becomes progressively more accurate and better captures the shape of the original signal.

When we resample the signal at $f_s=120\mathsf{Hz}$ the higher-frequency components (100 Hz and 200 Hz) are no longer represented correctly. This is a direct consequence of the Nyquist–Shannon sampling theorem, which states that the sampling frequency must be at least twice the highest frequency component in the signal to allow accurate reconstruction.

In our case, the original signal contains components at 100 Hz and 200 Hz, but the sampling rate is only 120 Hz. Since the Nyquist frequency is
$f_N = \tfrac{f_s}{2} = 60Hz$ both components exceed this limit and are therefore aliased into lower frequencies:

- The 100 Hz component is aliased to 20 Hz.
- The 200 Hz component is aliased to 40 Hz.

Alias frequencies can be computed as the absolute difference between the original frequency and the nearest integer multiple of the sampling frequency. You can read more [about aliasing here](https://brianmcfee.net/dstbook-site/content/ch02-sampling/Aliasing.html)).

### **Key takeaways**
- The resampled signal loses essential high-frequency information.
- Even worse, it introduces new frequency components (aliases) that were not present in the original signal.
- To prevent aliasing, a low-pass (anti-aliasing) filter must be applied before downsampling, ensuring that no frequency components above 
$\tfrac{f_s}{2}$ remain.
