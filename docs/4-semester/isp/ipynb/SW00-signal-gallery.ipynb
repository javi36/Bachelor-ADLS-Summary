## **Properties**

### **Periodicity**

A function $x(t)$ is periodic if there is a finite $T$ for which applies: $x(t+T) = x(t)$. The minimal $T$ for which this holds is called the *period T*. Sine and cosine have period $T=2\pi/\omega$.


Shift along the horizontal axis: $x_2(t) = x_1(t + c)$

### **Shift equivalence**

Sine and cosine differ only by a shift.

$$\begin{align*}
\sin(t + \tfrac{\pi}{2}) &= \cos(t)\\
\cos(t - \tfrac{\pi}{2}) &= \sin(t)
\end{align*}
$$

# **Convolution**

Convolution combines two signals, $f(t)$ and $g(t)$, according to the following definition:

$$(f*g)(t) = \int_\infty^\infty f(\tau)\cdot g(t-\tau)\cdot d\tau$$

The corresponding discrete formulation of the convolution is given by:

$$(f * g)[n] = \sum_{k=-\infty}^{\infty} f[k]\cdot g[n - k]$$

In this section, we visualize how the convolution is computed. The plots below are structured as follows:

- **Plot 1**: Input signal $f(\tau)$
- **Plot 2**: Input signal $g(\tau)$
- **Plot 3**: The product inside the integral $f(\tau)\cdot g(t-\tau)$, shown for increasing values of $t$
- **Plot 4**: The resulting convolution, i.e., the area under the product, evaluated for increasing values of $t$


## **Commutativity**

Convolution is **commutative**, meaning that the order of the signals does not affect the result:
$$(f * g)(t) = (g * f)(t)$$


The figures below illustrate this property by swapping the input functions $f(t)$ and $g(t)$.

## **Noisy signal vs. Hann window**

In this example, we filter a noisy signal using a Hann window. This type of filter acts as a smoothing filter, assigning higher weights to samples near the center and lower weights toward the edges. We can apply this filter using convolution, which produces a smoothed version of the original signal.

To avoid unintended scaling effects, we typically normalize the filter before applying the convolution. In general, when filtering a signal $x(t)$ with a filter $h(t)$, we do not want to change the overall scale or average value of the signal. Instead, we want to modify its structure, for example by smoothing or emphasizing certain features.

For this reason, it is common to normalize the filter such that its values sum to one — in other words, the area under the filter is one. This ensures that the output signal retains the same overall level as the input.

# **Cross-correlation**

Cross-correlation has a structure very similar to convolution. It combines two signals,  $f(t)$ and $g(t)$, according to the following definition:

$$(f \star g)(t) = \int_{-\infty}^{\infty} f(\tau)\cdot g(\tau - t)\, d\tau$$

The corresponding discrete formulation is given by:

$$(f \star g)[n] = \sum_{k=-\infty}^{\infty} f[k]\cdot g[k - n]$$

## **Commutativity**

Contrary to convolution, cross‑correlation is **not** commutative:

$$(g\star f)(t) \neq (f\star g)(t)$$

Indeed, for real-valued signals it holds that

$$(g\star f)(t) = (f\star g)(-t)$$

In words: Swapping the order of the signals in a cross‑correlation flips the time axis. That is, the correlation of $g$ with $g$ is the time‑reversed version of the correlation of $f$ with $g$.

## **Windowing**


<img src="../data/gallery/convolution/anim-convolution-hann-width.gif" alt="Effect of increasing window width on a noisy signal" width="600"/>

Observe the effect of a window filter on the noisy input signal:

- The larger the window, the more effectively it removes noise — but at the cost of smoothing sharp transitions (less steep edges).
- The smaller the window, the less effective the noise reduction — but the better it preserves sharp features.
