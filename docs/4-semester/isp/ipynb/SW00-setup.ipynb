## **Set up environment in Visual Studio Code**

It is recommended to create a virtual environment before installing the required Python packages. This ensures that the packages are installed in a clean environment and do not interfere with other projects.

Current versions of Visual Studio Code support virtual environments using either [`venv`](https://docs.python.org/3/library/venv.html) or [`anaconda`](https://www.anaconda.com/).

Follow the steps below to set up a new Jupyter Kernel in VS Code:

0. (Add the folder containing this Git repository to the workspace)
1. Open the command palette (Ctrl+Shift+P, Mac: Cmd+Shift+P)
2. Enter: Notebook: Select Notebook Kernel
3. Choose: Select Another Kernel
4. Choose: Python Environments
5. Choose: Create Python Environment
6. Choose: Venv or Conda
7. Choose the workspace directory containing this Git repository
8. Select an available Python installation (preferably Python 3.10 or newer)
9. A list of available requirements.txt files will be shown. Choose the one from this Git repository.
10. Hit Enter. The virtual environment is now created, and the kernel is ready to use.

After completing these steps, you will be able to select the kernel when opening a notebook in VS Code. The virtual environment will reside in the hidden `.venv` folder within the workspace directory you selected above.
